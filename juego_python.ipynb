```python
# EJERCICIO DE BONIFICACIÓN - Manejo de casos sin sentido en videojuegos

print("=== SISTEMA DE VALIDACIÓN DE VIDEOJUEGO ===\n")

# Caso 1: Vida negativa
print("--- Caso 1: Vida del jugador ---")
vida = int(input("Ingresa la vida del jugador: "))

if vida < 0:
    print("ERROR: La vida no puede ser negativa")
    vida = 0
    print(f"Vida corregida a: {vida}")
elif vida > 100:
    print("ERROR: La vida no puede superar 100")
    vida = 100
    print(f"Vida corregida a: {vida}")
else:
    print(f"Vida válida: {vida}")

print()

# Caso 2: Velocidad imposible
print("--- Caso 2: Velocidad del personaje ---")
velocidad = int(input("Ingresa la velocidad: "))

if velocidad < 0:
    print("ERROR: La velocidad no puede ser negativa")
    velocidad = 0
    print(f"Velocidad corregida a: {velocidad}")
elif velocidad > 200:
    print("ERROR: Velocidad demasiado alta (máximo 200)")
    velocidad = 200
    print(f"Velocidad corregida a: {velocidad}")
else:
    print(f"Velocidad válida: {velocidad}")

print()

# Caso 3: Nivel del jugador
print("--- Caso 3: Nivel del jugador ---")
nivel = int(input("Ingresa el nivel: "))

if nivel < 1:
    print("ERROR: El nivel mínimo es 1")
    nivel = 1
    print(f"Nivel corregido a: {nivel}")
elif nivel > 100:
    print("ERROR: El nivel máximo es 100")
    nivel = 100
    print(f"Nivel corregido a: {nivel}")
else:
    print(f"Nivel válido: {nivel}")

print()

# Caso 4: Cantidad de munición
print("--- Caso 4: Munición del arma ---")
municion = int(input("Ingresa la cantidad de munición: "))

if municion < 0:
    print("ERROR: No puedes tener munición negativa")
    municion = 0
    print(f"Munición corregida a: {municion}")
else:
    print(f"Munición válida: {municion}")

print()

# Caso 5: Coordenadas del mapa
print("--- Caso 5: Posición en el mapa ---")
x = int(input("Ingresa coordenada X: "))
y = int(input("Ingresa coordenada Y: "))

# Límites del mapa: 0 a 1000
if x < 0 or x > 1000:
    print(f"ERROR: X fuera de límites (0-1000)")
    if x < 0:
        x = 0
    else:
        x = 1000
    print(f"X corregida a: {x}")

if y < 0 or y > 1000:
    print(f"ERROR: Y fuera de límites (0-1000)")
    if y < 0:
        y = 0
    else:
        y = 1000
    print(f"Y corregida a: {y}")

print()


print("=== RESUMEN FINAL ===")
print(f"Vida: {vida}")
print(f"Velocidad: {velocidad}")
print(f"Nivel: {nivel}")
print(f"Munición: {municion}")
print(f"Posición: ({x}, {y})")
```
